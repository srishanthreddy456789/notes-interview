```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df1=df[df["Marks"] >80].sort_values(by="Marks")
print(df1)
```

```python
df["Class"].unique()
```

```python
x=[]
y=[]
z=[]
for i in range(0,df.shape[0]):
    if(df["Class"].iloc[i]=="X"):
        x.append(df["Marks"].iloc[i])
    if(df["Class"].iloc[i]=="Y"):
        y.append(df["Marks"].iloc[i])
    if(df["Class"].iloc[i]=="Z"):
        z.append(df["Marks"].iloc[i])
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
z_mean=np.mean(z)
x_mean,y_mean,z_mean
```

```python
sns.barplot(x='City',y='Marks',data=df)
plt.show()
```

```python
np.random.seed(42)
data=pd.DataFrame({
    "Category":np.repeat(["A","B","C"],30),
    "Values":np.concatenate([
        np.random.normal(10,2,30),
        np.random.normal(15,3,30),
        np.random.normal(20,4,30)
    ])
})
```

```python
sns.boxplot(x="Category",y="Values",data=data,notch=True)
plt.show()
```

```python
sns.boxplot(x="Values",y="Category",data=data)
plt.savefig("sea.png")
plt.show()
```
